```python
''' 
    calculates the basics for one Operator and one filename
'''

filename = "../output/HDF5_logfiles/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T24_L12_logfile.hdf5"
Op_ind = 0

(corr, ops, info) = HDF_log.get_corr_ops_info_from_HDF5_logfile(filename)
test = err.measurement("basic", measure_func=cf.basic_analysis, sampling_args = ("JK_SAMEDIM", 0, 0))
test.measure(orig_sample=np.swapaxes(np.mean(corr[Op_ind], axis=0),0,1), args=[filename, 0])
test.print_to_HDF()
test.print_everything()
```

```python
''' 
    calculates the basics for several operators and one filename
'''

filename = "../output/HDF5_logfiles/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T24_L12_logfile.hdf5"
Op_arr = [44,46,48]

(corr, ops, info) = HDF_log.get_corr_ops_info_from_HDF5_logfile(filename)
N_T = info[1]
corr_src = np.mean(corr,axis = 1)
print(len(corr_src),len(corr_src[0]),len(corr_src[0][0]))
data = []
args_0 = []
args_1 = N_T
t_mont = len(corr_src[0])
for i in range(t_mont):
    data.append([])
    for j in Op_arr:
        for k in range(len(corr_src[j][i])):
            data[i].append(corr_src[j][i][k])
print(len(data),len(data[0]))
for i in Op_arr:
    args_0.append(ops[i])

test = err.measurement("basic_all", measure_func=cf.basic_analysis_ops, sampling_args = ("JK_SAMEDIM", 0, 0))
test.measure(orig_sample=data, args=[args_0, args_1])
test.print_to_HDF()
test.print_everything()
```

```python
''' 
    calculates the basics for several operators and several filename
'''
Op_arr = [44,46,48]

print(os.listdir("../output/HDF5_logfiles"))

for name in os.listdir("../output/HDF5_logfiles"):
    filename = "../output/HDF5_logfiles/"+name
    (corr, ops, info) = HDF_log.get_corr_ops_info_from_HDF5_logfile(filename)
    N_L, N_T, gauge_group, beta, m_1, m_2 = info
    corr_src = np.mean(corr,axis = 1)
    print(len(corr_src),len(corr_src[0]),len(corr_src[0][0]))
    data = []
    args_0 = []
    args_1 = N_T
    t_mont = len(corr_src[0])
    for i in range(t_mont):
        data.append([])
        for j in Op_arr:
            for k in range(len(corr_src[j][i])):
                data[i].append(corr_src[j][i][k])
    print(len(data),len(data[0]))
    for i in Op_arr:
        args_0.append(ops[i])

    test = err.measurement("basic_%s_beta_%1.3f_m1_%1.3f_m2_%1.3f_T%i_L%i"%(gauge_group, beta, m_1, m_2, N_T, N_L), measure_func=cf.basic_analysis_ops, sampling_args = ("JK_SAMEDIM", 0, 0))
    test.measure(orig_sample=data, args=[args_0, args_1])
    test.print_to_HDF()
    test.print_everything()
```
