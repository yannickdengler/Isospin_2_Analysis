```python
data = []                                               # E_0s
args_0 = []                                             # N_Ls
args_1 = 0                                             # mass_Goldstone
stds = []

for filename in ("output/HDF5_corrfitter_results/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T16_L8_logfile.hdf5", "output/HDF5_corrfitter_results/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T20_L10_logfile.hdf5", "output/HDF5_corrfitter_results/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T24_L8_logfile.hdf5","output/HDF5_corrfitter_results/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T24_L12_logfile.hdf5","output/HDF5_corrfitter_results/Scattering_I2_SP(4)_beta6.900_m1-0.900_m2-0.900_T24_L14_logfile.hdf5"):
    with h5py.File("../"+filename,"r") as file:
        # file.visit(print)
        gauge_group = file["pi/gauge_group"][()].decode()
        beta = file["pi/beta"][()]
        m_1 = file["pi/m_1"][()]
        m_2 = file["pi/m_2"][()]
        N_T = file["pi/N_T"][()]
        N_L = file["pi/N_L"][()]
        E = file["pi/E"][()][0]
        E_err = file["pi/Delta_E"][()][0]

        data.append([E,])
        stds.append(E_err)
        args_0.append(N_L)
print(data)
print(stds)
print(args_0)
```

```python
test = err.measurement("m_inf_extrapolation_%s_beta_%1.3f_m1_%1.3f_m2_%1.3f"%(gauge_group, beta, m_1, m_2), measure_func=m_pi_inf.extrapolate_to_inf_mass, sampling_args = ("GAUSSIAN", stds, 100))
test.measure(orig_sample=data, args=[args_0, args_1])
test.print_to_HDF()
test.print_everything()
```
